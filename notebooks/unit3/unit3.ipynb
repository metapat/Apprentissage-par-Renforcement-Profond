## Entra√Ænez notre agent d'apprentissage par renforcement profond (Deep Q-Learning) √† jouer √† Space Invaders üëæ
Pour entra√Æner un agent avec RL-Baselines3-Zoo, il suffit de faire deux choses :

1. Cr√©ez un fichier de configuration d'hyperparam√®tres qui contiendra nos hyperparam√®tres d'entra√Ænement appel√©s dqn.yml.
Voici un exemple de mod√®le :

```
SpaceInvadersNoFrameskip-v4:
  env_wrapper:
    - stable_baselines3.common.atari_wrappers.AtariWrapper
  frame_stack: 4
  policy: 'CnnPolicy'
  n_timesteps: !!float 1e6
  buffer_size: 100000
  learning_rate: !!float 1e-4
  batch_size: 32
  learning_starts: 100000
  target_update_interval: 1000
  train_freq: 4
  gradient_steps: 1
  exploration_fraction: 0.1
  exploration_final_eps: 0.01
  # If True, you need to deactivate handle_timeout_termination
  # in the replay_buffer_kwargs
  optimize_memory_usage: False
```

Nous constatons ici que :
- Nous utilisons le Atari Wrapper pr√©traitement des donn√©es d'entr√©e (r√©duction du nombre d'images, conversion en niveaux de gris, empilement de 4 images).
- Nous utilisons CnnPolicydes couches convolutionnelles pour traiter les images, car nous les utilisons pour traiter les images.
- Nous l'entra√Ænons pour 10 millions n_timesteps
- La taille de la m√©moire (relecture de l'exp√©rience) est de 100 000, soit le nombre d'√©tapes d'exp√©rience que vous avez enregistr√©es pour entra√Æner √† nouveau votre agent.

üí° Je vous conseille de **r√©duire le nombre d'√©tapes d'entra√Ænement √† 1 million**, ce qui prendra environ 90 minutes sur un P100. `!nvidia-smi` La commande `git log` vous indiquera le GPU utilis√©. √Ä 10 millions d'√©tapes, cela prendra environ 9 heures. Je recommande d'ex√©cuter ce programme sur votre ordinateur (ou ailleurs). Cliquez simplement sur :  `File>Download`.

En mati√®re d'optimisation des hyperparam√®tres, je conseille de se concentrer sur ces 3 hyperparam√®tres :
- `learning_rate`
- `buffer_size (Experience Memory size)`
- `batch_size`

Il est recommand√© de consulter la documentation pour comprendre le r√¥le de chaque hyperparam√®tre : https://stable-baselines3.readthedocs.io/en/master/modules/dqn.html#parameters
